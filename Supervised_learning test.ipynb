# Régression simple

```python
X_reg1, Y_reg1 = make_regression(n_samples= 1000, n_features= 1, noise= 6)
print("Features shape : " + str(X_reg1.shape))
print("Target shape : " + str(Y_reg1.shape))

# Train Test

X_reg1_train, y_reg1_train, X_reg1_test, y_reg1_test = train_test_split(X_reg1, Y_reg1, test_size = 0.2, shuffle = True)
print("X_train shape : " + str(X_reg1_train.shape))
print("y_train shape : " + str(y_reg1_train.shape))
print("X_test shape : " + str(X_reg1_test.shape))
print("y_train shape : " + str(y_reg1_test.shape))
plt.figure()
plt.scatter(X_reg1, Y_reg1, s = 1)
plt.show()
```

```python
# Init models unsing Gradient descent
model_reg1_gd = Linear_Regression(n_iterations= 1000, gradient_descent=True,learning_rate = 0.01)
model_sk_reg1 = LinearRegression()

# Fit on the train set
model_reg1_gd.fit(X_reg1_train, y_reg1_train)
model_sk_reg1.fit(X_reg1_train, y_reg1_train)

# Score
print("R2 score mon modele : " + str(model_reg1_gd.get_score(X_reg1_test, y_reg1_test)))
print("R2 score sk model : " + str(model_sk_reg1.score(X_reg1_test, y_reg1_test)))

# Fit curve plot
model_reg1_gd.fit_curve()
```

```python
# Init models unsing Least square
model_reg1_ls = Linear_Regression(n_iterations= 1000, gradient_descent=False,learning_rate = 0.01)

# Fit on the train set
model_reg1_ls.fit(X_reg1_train, y_reg1_train)

# Score
print("R2 score mon modele : " + str(model_reg1_ls.get_score(X_reg1_test, y_reg1_test)))
print("R2 score sk model : " + str(model_sk_reg1.score(X_reg1_test, y_reg1_test)))
```

```python
res = model_reg1_ls.W[1] * np.arange(-5, 5) + model_reg1_ls.W[0]
plt.plot(np.arange(-5, 5), res, c ='r')
plt.scatter(X_reg1, Y_reg1, s = 1)
```

# Régression polynomiale simple

```python
X_reg3, Y_reg3 = make_regression(n_samples= 1000, n_features= 1, noise= 0)
Y_reg3 = X_reg3 ** 5 + 5*np.cos(X_reg3) + np.random.randn(1000,1)
Y_reg3 = Y_reg3.reshape((-1,))
X_reg3_train, y_reg3_train, X_reg3_test, y_reg3_test = train_test_split(X_reg3, Y_reg3, test_size = 0.2, shuffle = True)
print("X_train shape : " + str(X_reg3_train.shape))
print("y_train shape : " + str(y_reg3_train.shape))
print("X_test shape : " + str(X_reg3_test.shape))
print("y_train shape : " + str(y_reg3_test.shape))
plt.scatter(X_reg3,Y_reg3, s = 1)
```

```python
degree = 5
# Init models unsing Gradient descent
# Penser à augementer le nb_itérations et baisser le learning rate pour eviter le overflow uncountouring
model_reg3_gd = PolynomyaleRegression(degree= degree, n_iterations=1000, learning_rate=0.001)
model_sk_reg3 = LinearRegression()

# Fit on the train set
X_reg3_train_temp = polynomiale_features(X_reg3_train, degree=degree)
X_reg3_test_temp = polynomiale_features(X_reg3_test, degree=degree)
#poly = PolynomialFeatures(degree)
#X_reg3_train_temp = X_reg3_train
#X_reg3_test_temp = X_reg3_test
#X_reg3_train_temp = poly.fit_transform(X_reg3_train_temp)
#X_reg3_test_temp = poly.fit_transform(X_reg3_test_temp)

model_reg3_gd.fit(X_reg3_train, y_reg3_train)
model_sk_reg3.fit(X_reg3_train_temp, y_reg3_train)

# Get score
print("R2 score mon modele : " + str(model_reg3_gd.get_score(X_reg3_test, y_reg3_test)))
print("R2 score sk model : " + str(model_sk_reg3.score(X_reg3_test_temp, y_reg3_test)))

# Fit curve plot
model_reg3_gd.fit_curve()

# Predict
res_train = model_reg3_gd.predict(X_reg3)
plt.figure()
plt.scatter(X_reg3, Y_reg3, s = 4)
plt.scatter(X_reg3, res_train , s =2, c = 'r')
plt.show()
```

## Régressions pénalisées

```python
X_reg_pena, Y_reg_pena = make_regression(n_samples= 10000, n_features= 4, noise= 10)

Y_reg_pena = Y_reg_pena ** 5 + 5* np.cos(Y_reg_pena)
#Y_reg4 = Y_reg4.reshape((-1,))
X_reg_pena_train, y_reg_pena_train, X_reg_pena_test, y_reg_pena_test = train_test_split(X_reg_pena, Y_reg_pena, test_size = 0.2, shuffle = True)

deg = 5
X_reg_pena_train_temp = polynomiale_features(X_reg_pena_train, degree= deg)
X_reg_pena_test_temp = polynomiale_features(X_reg_pena_test, degree= deg)
print("X_train shape : " + str(X_reg_pena_train.shape))
print("y_train shape : " + str(y_reg_pena_train.shape))
print("X_test shape : " + str(X_reg_pena_test.shape))
print("y_train shape : " + str(y_reg_pena_test.shape))
```

```python
deg = 5
alpha = 0.5

model_lasso = LassoRegression(degree = deg, n_iterations=10000, learning_rate=0.01, alpha_reg= alpha)
model_sk_lasso = Lasso(alpha=alpha)

model_ridge = RidgeRegression(degree=deg, n_iterations=10000, learning_rate= 0.01, alpha_reg= alpha)
model_sk_ridge = Ridge(alpha= alpha)

model_elasticnet = Elastic_Net(degree = deg, n_iterations = 10000, learning_rate = 0.01, alpha_reg = alpha)
model_sk_elasticnet = ElasticNet(alpha = alpha)

model_lasso.fit(X_reg_pena_train, y_reg_pena_train)
model_sk_lasso.fit(X_reg_pena_train_temp, y_reg_pena_train)

model_ridge.fit(X_reg_pena_train, y_reg_pena_train)
model_sk_ridge.fit(X_reg_pena_train_temp, y_reg_pena_train)

model_elasticnet.fit(X_reg_pena_train, y_reg_pena_train)
model_sk_elasticnet.fit(X_reg_pena_train_temp, y_reg_pena_train)

print("------------------------")
print("R2 score mon lasso : " + str(model_lasso.get_score(X_reg_pena_test, y_reg_pena_test)))
print("R2 score mon ridge : " + str(model_ridge.get_score(X_reg_pena_test, y_reg_pena_test)))
print("R2 score mon elastic_net : " + str(model_elasticnet.get_score(X_reg_pena_test, y_reg_pena_test)))
print("------------------------")
print("R2 score sk lasso : " + str(model_sk_lasso.score(X_reg_pena_test_temp, y_reg_pena_test)))
print("R2 score sk ridge : " + str(model_sk_ridge.score(X_reg_pena_test_temp, y_reg_pena_test)))
print("R2 score sk elastic_net : " + str(model_sk_elasticnet.score(X_reg_pena_test_temp, y_reg_pena_test)))

```
